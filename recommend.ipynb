```python
ratings_name=ratings.merge(books, on ='ISBN')
```

```python
ratings_name
```

```python
num_rating_df = ratings_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
ratings_name['Book-Rating'] = pd.to_numeric(ratings_name['Book-Rating'], errors='coerce')
avg_rating_df = ratings_name.groupby('Book-Title', as_index=False)['Book-Rating'].mean()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_book_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_book_df
```

```python
popular_book_df=popular_book_df[popular_book_df['num_ratings']>=180].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_book_df
```

```python
popular_book_df=popular_book_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
popular_book_df
```

```python
x = ratings_name.groupby('User-ID').count()['Book-Rating']>200
avg_good_users = x[x].index.tolist()
```

```python
filtered_ratings = ratings_name[ratings_name['User-ID'].isin(avg_good_users)]
filtered_ratings
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
popular_books = y[y].index
popular_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(popular_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt.fillna(0,inplace=True)
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #fetching index:
    index = np.where(pt.index==book_name)[0][0]
    similar_book_items = [(index, float(score)) for index, score in sorted(
    list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True
)[1:6]]
    
    data = []
    
    for i in similar_book_items:
       
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data



    
```

```python
recommend('The Notebook')
```
